# Part B

```python
BTCdaily_returns = bitcoin_price.pct_change()*100
#uses pct_change function to create new dataframe from bitcoin_price
BTCdaily_returns.columns = ['pct_change']
#renaming the column from Closing Price to pct_change
BTCdaily_returns
#calling the dataframe
```

```python
BTCdaily_returns.plot(figsize=(20,10))
#plots daily returns

#the daily returns have huge variations
#this is indicates that there is no "best" time to invest into bitcoin, since it's so erratic int terms of return
```

Bitcoin's daily return has a dynamic change throughout each day; some days we can see the daily return was negative and other days there was a high positive return. This again implies the fact that Bitcoin does not have a guaranteed min daily return and it is very hard to predict the best time to buy or sell.
